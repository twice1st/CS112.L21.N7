# **BÀI 1: Đổi mới**

1. **Abstraction**: Số cây đốn hạ của 2 đội lần lượt là a và b, ngày nghỉ của 2 đội lần lượt là thứ k và thứ m. Tìm số ngày sau khi đốn hết n cây.
2. **Recognition**: Sử dụng thuật toán Binary Search.
3. **Algorithm designed**: (Lưu đồ thuật toán)
4. **Alogrithm Complexity**: log(n)


image.png

5. **Programming**:

```python
a,k,b,m,n = [int(i) for i in input().split()]
low = n//(a+b)
high = n
while low <= high:
    mid = (low+high)//2
    if a*(mid-mid//k)+ b*(mid-mid//m) >= n:
        day = mid
        high = mid - 1
    else:
        low = mid + 1
print(day)
```

# **BÀI 2: Số nguyên mới**

1. **Abstraction**: Cho 0 < n <= 10^99, tìm số nguyên lớn nhất m là số chia hết cho 3 và khác n ở đúng 1 chữ số.
2. **Recognition**: Dùng tính chất số dư khi chia cho 3.
3. **Algorithm designed**: 
> 1. *Nhập số n và tạo ls là list gồm các số 9, 8, 7 tương ứng với index 0, 1, 2 là các số dư khi chia cho 3*
>
> 2. *Tạo ls_num là list các chữ số của số n*
>
> 3. *Vòng lặp duyệt qua các chữ số của n*
>       
>       `for i in range(len(ls_num)):`
>   
>       *Tính tổng sum của các chữ số còn lại của n :*
>       
>           sum_rest_num = sum(ls_num)-ls_num[i]
>       *Tính số dư của sum_rest_num khi chia cho 3*
>           t = (sum_rest_num) mod 3
>       *If chữ số vòng lặp đang xét bé hơn ls[t]:*
>           Chữ số đang xét = ls[t]
>           Thoát vòng lặp
>
>4. *If số n_ mới không chia hết cho 3 hoặc bằng số n ban đầu:*
>           n_ = n_ - 1
>           Quay lại 4
>5. *In kết quả ra màn hình*
>
4. **Alogrithm Complexity**: O(n)


5. **Programming:**

```python
n = input()
ls = [9, 8, 7]
ls_num = list(map(int, list(n)))
for i in range(len(ls_num)):
    sum_rest_num = sum(ls_num)-ls_num[i]
    t = ls[sum_rest_num % 3]
    if ls_num[i] < t:
        ls_num[i] = t
        break

kq = int(''.join(list(map(str,ls_num))))


while kq % 3 !=0 or kq == int(n):
    kq -= 1

print(kq)
```
